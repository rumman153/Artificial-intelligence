```python
import pandas as pd
df = pd.read_csv('/content/gdrive/MyDrive/Semester_4.1/CSE_404/assignment_2/roadmap_kg_UAP.csv') #reading the csv file
df
```

```python
initial_state = 'kishoreganj' #assigning start node
goal = 'UAP' #assigning goal node
kg_hn = 3.59 #assigning the huerestic value of start node kishoreganj
df.columns = [ 'Parent', 'Child', 'g(n)', 'h(n)'] #assigning the headers of the columns
df
```

```python
df['g(n)'] = df['g(n)'].astype('float32') #assigning variable type
df['h(n)'] = df['h(n)'].astype('float32') #assigning variable type
path = [initial_state] #declaring a list
sdf = df.copy() #making a copy 
path_cost = 0 #declaring a variable
while True: #starting a loop
    if initial_state==goal: #checking if found the goal
        break
    sd = sdf[sdf.Parent== initial_state] #putting header value as parent 
    sd['f(n)']=sd[['g(n)','h(n)']].sum(axis=1) #calculating the f(n) value
    
    sd=sd.sort_values(by='f(n)',ascending=True).reset_index(drop=True) #sorting the rows in ascending order
    print(sd) #printing the rows
    initial_state = sd.loc[0].Child #setting the new header
    path_cost+=sd.loc[0]['g(n)']#.value #calculating the path cost
    path_cost+=sd.loc[0]['h(n)']
    path.append(initial_state) #putting the new header into the list
```

```python
print('Optimal path ',path) #printing the optimal path
print('Optimal path cost is ',kg_hn+path_cost,'km') #printing the path cost
```
